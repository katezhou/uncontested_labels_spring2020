```python
from numpy.random import choice
from collections import Counter
```

```python
for sample_size in [1, 3, 5, 7]:
    count = 0
    for iteration in range(0,100):
        if(iteration % 10 == 0):
            print(iteration)
        for x in mturk.image_fraction_selected.keys():
            list_of_candidates = []
            probabilities = []
            for y in mturk.image_fraction_selected[x]:
                #thresholding
                if (mturk.image_fraction_selected[x][y] > 0.7):
                    list_of_candidates.append(y)
                    probabilities.append(mturk.image_fraction_selected[x][y])
            for i, x in enumerate(list_of_candidates):
                random = choice([0, 1], sample_size * 2, p=[1-probabilities[i], probabilities[i]], replace=True)
                count1 = Counter(random[:sample_size]).most_common(2)
                count2 = Counter(random[sample_size:]).most_common(2)
                majority_1 = count1[0][0]
                majority_2 = count2[0][0]  
                if ((len(count1) > 1 and count1[0][1] == count1[1][1])): #no majority
                    majority_1 = -2
                if (len(count2) > 1 and count2[0][1] == count2[1][1]): #no majority
                    majority_2 = -2
                if (majority_1 != majority_2):
                    count += 1
    print('percent', count/49339)
```

```python
count = 0
for iteration in range(0,100):
    if(iteration % 10 == 0):
        print(iteration)
    for x in mturk.image_fraction_selected.keys():
        total = 0.0
        list_of_candidates = []
        probability_distribution = []
        for y in mturk.image_fraction_selected[x]:
            total += mturk.image_fraction_selected[x][y]
            list_of_candidates.append(y)
            probability_distribution.append(mturk.image_fraction_selected[x][y])
        if (total > 0.0):
            probability_distribution = [x / sum(probability_distribution) for x in probability_distribution]
            random = choice(list_of_candidates, 6, p=probability_distribution, replace=True)
            count1 = Counter(random[:3]).most_common(1)
            count2 = Counter(random[3:]).most_common(1)
            if (count1[0][1] < 2 or count2[0][1] < 2 ): #no majority - flip
                count += 1
            elif (count1[0][0] !=  count2[0][0] ): #different flip
                count += 1
```
